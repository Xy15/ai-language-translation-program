```python
import urllib.request
import easygui
import re
import string
import os
import textblob
import argparse

from googletrans import Translator
from bs4 import BeautifulSoup
from nltk.stem import WordNetLemmatizer
from gtts import gTTS
from textblob import TextBlob

from pytesseract import TesseractError

# Import necessary libraries for image
import cv2
import pytesseract
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
from pytesseract import Output
from matplotlib.image import imread
from imutils.perspective import four_point_transform
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

translator = Translator()


title = "Translation Program"
transChoice = ["Image","Website","File"]
transButtons = easygui.buttonbox("Choose type to translate", title, transChoice)

try:
    if transButtons == transChoice[0]:
        imgTitle = "Image Choice"
        imgChoice = ["Upload","Scan"]
        imgButtons = easygui.buttonbox("Choose type to translate", imgTitle, imgChoice)
        
        if imgButtons == imgChoice[0]:
            # Create a file open dialog using EasyGUI
            file_path = easygui.fileopenbox(msg="Select an image file", filetypes=["*.jpg", "*.png", "*.bmp"])
            # Check if the selected file is an image
            # Load the image using Matplotlib
            img = cv2.imread(file_path)
            # Display the image using Matplotlib
            cv2.imshow('image', img)
            cv2.waitKey(0)
            cv2.destroyAllWindows()
            
        elif imgButtons == imgChoice[1]:
            cap = cv2.VideoCapture(0 + cv2.CAP_DSHOW)
            WIDTH, HEIGHT = 800, 600
            cap.set(cv2.CAP_PROP_FRAME_WIDTH, WIDTH)
            cap.set(cv2.CAP_PROP_FRAME_HEIGHT, HEIGHT)

            document_contour = np.array

            def scan_detection(image):
                global document_contour

                document_contour = np.array([[0, 0], [WIDTH, 0], [WIDTH, HEIGHT], [0, HEIGHT]])

                gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                blur = cv2.GaussianBlur(gray,(5,5),0)
                threshold = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
                contours, _ = cv2.findContours(threshold, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
                contours = sorted(contours, key=cv2.contourArea, reverse=True)

                max_area = 0
                for contour in contours:
                    area = cv2.contourArea(contour)
                    if area > 1000:
                        peri = cv2.arcLength(contour, True)
                        approx = cv2.approxPolyDP(contour, 0.015 * peri, True)
                        if area > max_area and len(approx) == 4:
                            document_contour = approx
                            max_area = area
                cv2.drawContours(frame, [document_contour], -1, (0, 255, 0), 3)

            while True:
                _, frame = cap.read()
                frame_copy = frame.copy()

                scan_detection(frame_copy)

                cv2.imshow("input", frame)
                warped = four_point_transform(frame_copy, document_contour.reshape(4,2))

                if cv2.waitKey(1) & 0xFF == 27: #Captured 'Esc' key input
                    break

            cv2.imshow("warped", warped)

            while True:

                if cv2.waitKey(1) & 0xFF == 13:  #Captured 'Enter' key input
                    cv2.imwrite('save.jpg', warped)  # save warped image
                    print('saved')
                    break
                elif cv2.waitKey(1) & 0xFF == 27:
                    break

            img = warped
            print(pytesseract.image_to_string(warped))
            cap.release()  # Turn off camera
            cv2.destroyAllWindows()
        sourceTitle = "Language in image"
        choices = ["Chinese", "English"]
        sourceLang = easygui.choicebox("Please choose image file language", sourceTitle, choices)
        if sourceLang == "Chinese":
            sLang = 'chi_sim'
        elif sourceLang == "English":
            sLang = 'eng'
        
        # Resize
        h, w = img.shape[:2]
        if (w > 800 or h > 800):
            w = w/2
            h = h/2
        elif (w < 300 or h < 300):
            w *= 2
            h *= 2

        w = int(w / 32) * 32
        h = int(h / 32) * 32

        img = cv2.resize(img, (w, h))
        cv2.imshow('Resized',  img)
        cv2.waitKey(0)
        cv2.destroyAllWindows() 
        
        # Normalization - improve the image's contrast, brightness,
        norm_img = np.zeros((img.shape[0], img.shape[1]))
        normalize_img = cv2.normalize(img, norm_img, 0, 255, cv2.NORM_MINMAX)

        # Convert to grayscale
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        normalize_gray = cv2.cvtColor(normalize_img,cv2.COLOR_BGR2GRAY)

        names = ['Original Image','Normalization','Grayscale', 'Normalize Gray']
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        normalize_img = cv2.cvtColor(normalize_img, cv2.COLOR_BGR2RGB)
        images = img,normalize_img,gray, normalize_gray

        for i in range(4):
            plt.subplot(2,2,i+1),plt.imshow(images[i], 'gray')
            plt.title(names[i])
            plt.xticks([]),plt.yticks([])

        plt.show()
        cv2.imshow('img', img)
        cv2.imshow('normalize_img', normalize_img)
        cv2.imshow('gray', gray)
        cv2.imshow('normalize_gray', normalize_gray)
        img = normalize_gray
        print(pytesseract.image_to_string(img, sLang))
        cv2.waitKey(0)
        cv2.destroyAllWindows()    

        # Create the sharpening kernel
        kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
        # Apply the sharpening kernel to the image using filter2D
        sharpened = cv2.filter2D(img, -1, kernel)

        cv2.imshow('img', img)
        cv2.imshow('sharpened', sharpened)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

        img = sharpened

        def rotate(image, inverse): 
            if(inverse):
                img = cv2.bitwise_not(image)
            else:
                img = image
            blur = cv2.GaussianBlur(img,(5,5),0)
            # threshold the image, all foreground pixels: 255, all background pixels to 0
            thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]

            # grab the (x, y) coordinates of all pixel values that are greater than zero, 
            # compute a bounding box that contains all coordinates
            coords = np.column_stack(np.where(thresh > 0))

            # calculate the angle of rotation of the minimum area rectangle 
            angle = cv2.minAreaRect(coords)[-1]

            # the `cv2.minAreaRect` function returns values in the range [-90, 0)
            # if rotated more than 90 degrees, the next edge is used to calculate the angle from the horizontal
            if angle < -45:
                angle = -(90 + angle)
            # take the inverse of the angle to make it positive
            else:
                angle = -angle

            # rotate the image to deskew it
            (h, w) = image.shape[:2]
            center = (w // 2, h // 2)
            M = cv2.getRotationMatrix2D(center, angle, 1.0)
            rotated = cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_CUBIC, borderMode=cv2.BORDER_REPLICATE)
            rotatedAngle = rotated.copy()
            print(rotatedAngle)
            # draw the correction angle on the image so we can validate it
            cv2.putText(rotatedAngle, "Angle: {:.2f} degrees".format(angle), (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
            # show the output image
            print("Angle: {:.2f}".format(angle))

            cv2.imshow("Input", image)
            cv2.imshow("Rotated", rotatedAngle)
            cv2.waitKey(0)
            cv2.destroyAllWindows()
            return rotated
        
        rotateTitle = "Rotate Image"
        rotateChoice = ["BLACK text on WHITE background ","WHITE text on DARK background", "Rotated 180 degrees", "Other"]
        rotateButtons = easygui.buttonbox("The rotation result may not be accurate if the image does not have a clear contrast between the text and the background, such as black text on a white background or white text on a dark background.\nPlease select the type of image: ", rotateTitle, rotateChoice)
        if rotateButtons == rotateChoice[1]:
            rotated = rotate(img, 0)
        elif rotateButtons == rotateChoice[2]:
            rotated = cv2.rotate(img, cv2.ROTATE_180)
        else:
            # flip the foreground and background to ensure foreground is now "white" and the background is "black"
            rotated = rotate(img, 1) #inverts pixel values
            
        
        rotateTitle = "Rotate Image"
        rotateChoice = ["Yes","No"]
        rotateButtons = easygui.buttonbox("Accept rotation? ", rotateTitle, rotateChoice)
        if rotateButtons == rotateChoice[0]:
            img = rotated

        # Noise Removal
        # Averaging filtering
        blurAverage = cv2.blur(img,(5,5))

        # Gaussian filtering
        blurGauss = cv2.GaussianBlur(img,(5,5),0)

        # Median filtering
        blurMedian = cv2.medianBlur(img,5)

        # non-local means denoising
        nonLocalMeans = cv2.fastNlMeansDenoising(img, None, 10, 10, 7)

        names = ['picBlur','blurGauss','blurMedian','nonLocalMeans']
        images = blurAverage,blurGauss,blurMedian,nonLocalMeans

        for i in range(4):
            plt.subplot(2,2,i+1),plt.imshow(images[i], 'gray')
            plt.title(names[i])
            plt.xticks([]),plt.yticks([])

        plt.show()
        
        cv2.imshow('img', img)
        cv2.imshow('blurAverage', blurAverage)
        cv2.imshow('blurGauss', blurGauss)
        cv2.imshow('blurMedian', blurMedian)
        cv2.imshow('nonLocalMeans', nonLocalMeans)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

        imgT = "Noise Removal"
        imgC = ["Original", "Blur Average", "Blur Gaussian", "Blur Median", "Non Local Means"]
        ch = easygui.buttonbox("From the text displayed, choose one that is closest to the text in the image." + "\nOriginal: \n" + pytesseract.image_to_string(img, sLang)+ "\nBlur Average: \n" + pytesseract.image_to_string(blurAverage, sLang)+ "\nBlur Gauss: \n" + pytesseract.image_to_string(blurGauss, sLang)+ "\nBlur Median: \n" + pytesseract.image_to_string(blurMedian, sLang)+ "\nNon Local Means: \n"+ pytesseract.image_to_string(nonLocalMeans, sLang), imgT, imgC)
        if ch == imgC[0]:
            img = img
        elif ch == imgC[1]:
            img = blurAverage
        elif ch == imgC[2]:
            img = blurGauss
        elif ch == imgC[3]:
            img = blurMedian
        elif ch == imgC[4]:
            img = nonLocalMeans

        # Erosion
        kernel = np.ones((3,3),np.uint8) #To define the structuring element
        erosion = cv2.erode(img,kernel,iterations = 1)

        # Dilation
        kernel = np.ones((5,5),np.uint8)
        dilation = cv2.dilate(img,kernel,iterations = 1)

        # Opening
        kernel = np.ones((3,3),np.uint8)
        opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

        # Closing
        kernel = np.ones((5,5),np.uint8)
        closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

        # Canny Edge Detection
        canny = cv2.Canny(img, 100, 200)

        names = ['Original','Erosion','Dilation','Opening','Closing','Canny']
        images = img,erosion,dilation,opening,closing,canny

        for i in range(5):
            plt.subplot(2,3,i+1),plt.imshow(images[i], 'gray')
            plt.title(names[i])
            plt.xticks([]),plt.yticks([])
        plt.show()
        
        cv2.imshow('img', img)
        cv2.imshow('erosion', erosion)
        cv2.imshow('dilation', dilation)
        cv2.imshow('opening', opening)
        cv2.imshow('closing', closing)
        cv2.imshow('canny', canny)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

        imgT = "Morphological Operations"
        ch = easygui.buttonbox("From the text displayed, choose one that is closest to the text in the image." + "\nOriginal \n" + pytesseract.image_to_string(img, sLang) + "\nErosion: \n" + pytesseract.image_to_string(erosion, sLang)+ "\nDilation: \n" + pytesseract.image_to_string(dilation, sLang)+ "\nOpening: \n" + pytesseract.image_to_string(opening, sLang)+ "\nClosing: \n" + pytesseract.image_to_string(closing, sLang)+ "\nCanny: \n"+ pytesseract.image_to_string(canny, sLang), imgT, names)
        if ch == names[0]:
            img = img
        elif ch == names[1]:
            img = erosion
        elif ch == names[2]:
            img = dilation
        elif ch == names[3]:
            img = opening
        elif ch == names[4]:
            img = closing
        elif ch == names[5]:
            img = canny
            
        # Binarization/Thresholding
        gray = img

        ret,thresh_binary = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
        ret,thresh_binary_inv = cv2.threshold(gray,127,255,cv2.THRESH_BINARY_INV)
        ret,thresh_trunc = cv2.threshold(gray,127,255,cv2.THRESH_TRUNC)
        ret,thresh_tozero = cv2.threshold(gray,127,255,cv2.THRESH_TOZERO)
        ret,thresh_tozero_inv = cv2.threshold(gray,127,255,cv2.THRESH_TOZERO_INV)
        thresh_otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

        #Adaptive Threshold: different threshold values are used for different parts of the image. This function gives better results for images with varying lighting conditions
        #here 11 is the pixel neighbourhood that is used to calculate the threshold value
        thresh_mean = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
        thresh_gaussian = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)

        names = ['Original Image','BINARY','THRESH_BINARY_INV','THRESH_TRUNC','THRESH_TOZERO','THRESH_TOZERO_INV','ThreshOtsu', 'Adaptive Mean','Adaptive Gaussian']
        images = gray,thresh_binary,thresh_binary_inv,thresh_trunc,thresh_tozero,thresh_tozero_inv, thresh_otsu, thresh_mean, thresh_gaussian

        for i in range(9):
            plt.subplot(3,3,i+1),plt.imshow(images[i], 'gray')
            plt.title(names[i])
            plt.xticks([]),plt.yticks([])    
        plt.show()
        
        imgT = "Thresholding"
        imgC = ["binary", "thresh_trunc", "thresh_tozero", "thresh_otsu", "thresh_mean", "thresh_gaussian"]
        ch = easygui.buttonbox("From the text displayed, choose one that is closest to the text in the image." + 
                               "\nOriginal Image: \n" + pytesseract.image_to_string(gray, sLang)+ 
                               "\nBinary Thresh: \n" + pytesseract.image_to_string(thresh_binary, sLang)+ 
                               "\nThresh_Trunc: \n" + pytesseract.image_to_string(thresh_trunc, sLang)+ 
                               "\nThresh_tozero: \n" + pytesseract.image_to_string(thresh_tozero, sLang)+ 
                               "\nThresh Otsu: \n" + pytesseract.image_to_string(thresh_otsu, sLang)+ 
                               "\nThresh_mean: \n" + pytesseract.image_to_string(thresh_mean, sLang)+ 
                               "\nThresh_gaussian: \n" + pytesseract.image_to_string(thresh_gaussian, sLang), imgT, imgC)

        if ch == imgC[0]:
            img = gray
        elif ch == imgC[1]:
            img = thresh_binary
        elif ch == imgC[2]:
            img = thresh_trunc
        elif ch == imgC[3]:
            img = thresh_tozero
        elif ch == imgC[4]:
            img = thresh_otsu
        elif ch == imgC[5]:
            img = thresh_mean
        elif ch == imgC[6]:
            img = thresh_gaussian

        #OCR
        def boxes_character(image):
            # Boxes around each character
            text = ''
            hImg, wImg = image.shape
            boxes = pytesseract.image_to_boxes(image, lang=sLang)
            for b in boxes.splitlines():
                b = b.split(' ')
                print(b)
                x, y, w, h = int(b[1]), int(b[2]), int(b[3]), int(b[4])
                # Draw Box around text: cv2.rectangle(‘image_name’, x_coordinate, y_coordinate, RGB_value_of_color, thickness_of_box)
                cv2.rectangle(image, (x, hImg - y), (w, hImg - h), (50, 50, 255), 1)
                # Label: cv2.putText(‘image_name’, x_coordinate, y_coordinate, ‘font_name’, font_size, RGB_value_of_color, thickness_of_text)
                cv2.putText(image, b[0], (x, hImg - y + 13), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (50, 205, 50), 1)
                text += b[0]

            return text

        def boxes_text(image):
            # Boxes around each text
            d = pytesseract.image_to_data(image, output_type=Output.DICT, lang=sLang)

            # initialize
            joined_text = ""
            prev_x, prev_y, prev_w = 0, 0, 0

            n_boxes = len(d['text'])
            for i in range(n_boxes):
                if int(d['conf'][i]) > 60:
                    word = d['text'][i]
                    (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])   #(x,y) -> top-left corner's coordinate,w=width, h=height
                    image = cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

                    # Check horizontal distance between current and previous box
                    if (x - prev_x - prev_w) < h/2:
                        joined_text += " "
                    else:
                        joined_text += "\n"

                    joined_text += word

                    # Store previous box information for next iteration
                    prev_x, prev_y, prev_w = x, y, w
            return joined_text

        boxesChar = img.copy()
        boxesTxt = img.copy()

        text = boxes_character(boxesChar)
        joined_text = boxes_text(boxesTxt)
        print(text)
        print(joined_text)

        cv2.imshow('Boxes Character', boxesChar)
        cv2.imshow('Boxes Text', boxesTxt)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

        imgT = "Detected Text"
        imgC = ["Option 1", "Option 2", "Option 3"]
        ch = easygui.buttonbox("From the text displayed, choose one that is closest to the text in the image." + "\nOption 1: \n" + text+ "\nOption 2: \n" + joined_text + "\nOption 3: \n" + pytesseract.image_to_string(img), imgT, imgC)
        if ch == imgC[0]:
            transText = text
        elif ch == imgC[1]:
            transText = joined_text
        elif ch == imgC[2]:
            transText = pytesseract.image_to_string(img,lang=sLang)

    elif transButtons == transChoice[1]:
        title = "Open Website"
        url =  easygui.enterbox("Please enter website URL", title)
        import urllib.request
        with urllib.request.urlopen(url) as f:
            rawHTML = f.read()
        transText = BeautifulSoup(rawHTML).get_text()
    elif transButtons == transChoice[2]:
        title = "Open File"
        directory = easygui.fileopenbox("Please enter file directory", title)
        with open(directory, "r", encoding="utf-8") as f:
            transText = f.read()


    chTitle = "Translate or Fix"
    tChoice = ["Translate", "Fix Typo"]
    btnChoice = easygui.buttonbox("Choose one of the following", chTitle, tChoice)
    if btnChoice == tChoice[0]:
        s = re.sub('[^\w\s]','',transText)
        a = s[:]

        choices = ["Afrikaans", "Albanian", "Amharic", "Arabic", "Armenian", "Assamese", "Aymara", "Azerbaijani", "Bambara", "Basque", "Belarusian", "Bengali", "Bhojpuri", "Bosnian", "Bulgarian", "Catalan", "Cebuano", "Chinese (Simplified)", "Chinese (Traditional)", "Corsican", "Croatian", "Czech", "Danish", "Dhivehi", "Dogri", "Dutch", "English", "Esperanto", "Estonian", "Ewe", "Filipino (Tagalog)", "Finnish", "French", "Frisian", "Galician", "Georgian", "German", "Greek", "Guarani", "Gujarati", "Haitian Creole", "Hausa", "Hawaiian", "Hebrew", "Hindi", "Hmong", "Hungarian", "Icelandic", "Igbo", "Ilocano", "Indonesian", "Irish", "Italian", "Japanese", "Javanese", "Kannada", "Kazakh", "Khmer", "Kinyarwanda", "Konkani", "Korean", "Krio", "Kurdish", "Kurdish (Sorani)", "Kyrgyz", "Lao", "Latin", "Latvian", "Lingala", "Lithuanian", "Luganda", "Luxembourgish", "Macedonian", "Maithili", "Malagasy", "Malay", "Malayalam", "Maltese", "Maori", "Marathi", "Meiteilon (Manipuri)", "Mizo", "Mongolian", "Myanmar (Burmese)", "Nepali", "Norwegian", "Nyanja (Chichewa)", "Odia (Oriya)", "Oromo", "Pashto", "Persian", "Polish", "Portuguese (Portugal, Brazil)", "Punjabi", "Quechua", "Romanian", "Russian", "Samoan", "Sanskrit", "Scots Gaelic", "Sepedi", "Serbian", "Sesotho", "Shona", "Sindhi", "Sinhala (Sinhalese)", "Slovak", "Slovenian", "Somali", "Spanish", "Sundanese", "Swahili", "Swedish", "Tagalog (Filipino)", "Tajik", "Tamil", "Tatar", "Telugu", "Thai", "Tigrinya", "Tsonga", "Turkish", "Turkmen", "Twi (Akan)", "Ukrainian", "Urdu", "Uyghur", "Uzbek", "Vietnamese", "Welsh", "Xhosa", "Yiddish", "Yoruba", "Zulu"]
        langCode = ["af", "sq", "am", "ar", "hy", "as", "ay", "az", "bm", "eu", "be", "bn", "bho", "bs", "bg", "ca", "ceb", "zh-CN", "zh-TW", "co", "hr", "cs", "da", "dv", "doi", "nl", "en", "eo", "et", "ee", "fil", "fi", "fr", "fy", "gl", "ka", "de", "el", "gn", "gu", "ht", "ha", "haw", "he", "hi", "hmn", "hu", "is", "ig", "ilo", "id", "ga", "it", "ja", "jv", "kn", "kk", "km", "rw", "gom", "ko", "kri", "ku", "ckb", "ky", "lo", "la", "lv", "ln", "lt", "lg", "lb", "mk", "mai", "mg", "ms", "ml", "mt", "mi", "mr", "mni-Mtei", "lus", "mn", "my", "ne", "no", "ny", "or", "om", "ps", "fa", "pl", "pt", "pa", "qu", "ro", "ru", "sm", "sa", "gd", "nso", "sr", "st", "sn", "sd", "si", "sk", "sl", "so", "es", "su", "sw", "sv", "tl", "tg", "ta", "tt", "te", "th", "ti", "ts", "tr", "tk", "ak", "uk", "ur", "ug", "uz", "vi", "cy", "xh", "yi", "yo", "zu"]    

        language = easygui.choicebox("Please choose a language", title, choices)
        ch = len(choices)

        for i in range(ch):
            if language == choices[i]:
                l = langCode[i]
                break
        if transButtons == transChoice[0]:
            easygui.msgbox(translator.translate(a, dest=l).text)
        else: 
            easygui.msgbox(translator.translate(a, dest=l).text)

        ttsTitle = "Text To Speech?"
        ttsChoice = ["Yes", "No"]
        ttsButtons = easygui.buttonbox("Do you want text to speech?", ttsTitle, ttsChoice)

        if ttsButtons == ttsChoice[0]:
            translated = translator.translate(a, dest=l).text
            try:
                audio = gTTS(text=translated, lang=l, slow=False)
                audio.save("translated.mp3")
                os.system("start translated.mp3")
            except:
                print("Language not supported")
    elif btnChoice == tChoice[1]:
        s = re.sub('[^\w\s]','',transText)
        text = TextBlob(s)
        fix = text.correct()
        easygui.msgbox(fix)
except: 
     easygui.msgbox("An error has occured.\nErrors may be:\n1. File type not supported\n2. Language Not Supported\n3. Website not supported\n\nPlease do note that HTTPS websites are currently not supported, same goes to files other than .txt files")
```
